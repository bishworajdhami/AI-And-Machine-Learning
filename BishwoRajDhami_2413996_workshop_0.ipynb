### Task 1 — Unit Converter

```python
def convert_length(value, direction):
    """
    Converts length between meters and feet.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'mf' for meters->feet, 'fm' for feet->meters.

    Returns:
        float: The converted length value.
    """
    if direction == 'mf':
        return value * 3.28084
    elif direction == 'fm':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction. Use 'mf' or 'fm'.")


def convert_weight(value, direction):
    """
    Converts weight between kilograms and pounds.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'kl' for kg->lbs, 'lk' for lbs->kg.

    Returns:
        float: The converted weight value.
    """
    if direction == 'kl':
        return value * 2.20462
    elif direction == 'lk':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction. Use 'kl' or 'lk'.")


def convert_volume(value, direction):
    """
    Converts volume between liters and gallons.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'lg' for liters->gallons, 'gl' for gallons->liters.

    Returns:
        float: The converted volume value.
    """
    if direction == 'lg':
        return value * 0.264172
    elif direction == 'gl':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction. Use 'lg' or 'gl'.")


def unit_converter():
    """
    Interactive unit converter for length, weight, and volume.
    Prompts user for conversion type and value, then displays the result.
    Handles invalid input using try-except.

    Returns:
        float or None: The converted result, or None on invalid input.
    """
    print("=== Unit Converter ===")
    print("1. Length  (meters <-> feet)")
    print("2. Weight  (kilograms <-> pounds)")
    print("3. Volume  (liters <-> gallons)")

    try:
        category = input("Choose conversion type (1/2/3): ").strip()

        if category == '1':
            print("  a. Meters to Feet")
            print("  b. Feet to Meters")
            sub = input("Choose direction (a/b): ").strip().lower()
            direction = 'mf' if sub == 'a' else 'fm'
            value = float(input("Enter value: "))
            result = convert_length(value, direction)
            unit_from = 'm' if sub == 'a' else 'ft'
            unit_to   = 'ft' if sub == 'a' else 'm'

        elif category == '2':
            print("  a. Kilograms to Pounds")
            print("  b. Pounds to Kilograms")
            sub = input("Choose direction (a/b): ").strip().lower()
            direction = 'kl' if sub == 'a' else 'lk'
            value = float(input("Enter value: "))
            result = convert_weight(value, direction)
            unit_from = 'kg' if sub == 'a' else 'lbs'
            unit_to   = 'lbs' if sub == 'a' else 'kg'

        elif category == '3':
            print("  a. Liters to Gallons")
            print("  b. Gallons to Liters")
            sub = input("Choose direction (a/b): ").strip().lower()
            direction = 'lg' if sub == 'a' else 'gl'
            value = float(input("Enter value: "))
            result = convert_volume(value, direction)
            unit_from = 'L' if sub == 'a' else 'gal'
            unit_to   = 'gal' if sub == 'a' else 'L'

        else:
            print("Invalid category. Please choose 1, 2, or 3.")
            return None

        print(f"\nResult: {value} {unit_from} = {result:.4f} {unit_to}")
        return result

    except ValueError as e:
        print(f"Error: {e}. Please enter a valid numeric value.")
        return None



unit_converter()
```

### Task 2 — Mathematical Operations on a List of Numbers

```python
def compute_sum(numbers):
    """
    Computes the sum of a list of numbers.

    Parameters:
        numbers (list of float): Input list of numeric values.

    Returns:
        float: The sum of all elements in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot compute sum of an empty list.")
    total = 0
    for n in numbers:
        total += n
    return total


def compute_average(numbers):
    """
    Computes the average of a list of numbers.

    Parameters:
        numbers (list of float): Input list of numeric values.

    Returns:
        float: The arithmetic mean of the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot compute average of an empty list.")
    return compute_sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
        numbers (list of float): Input list of numeric values.

    Returns:
        float: The largest element in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot find maximum of an empty list.")
    max_val = numbers[0]
    for n in numbers[1:]:
        if n > max_val:
            max_val = n
    return max_val


def compute_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
        numbers (list of float): Input list of numeric values.

    Returns:
        float: The smallest element in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot find minimum of an empty list.")
    min_val = numbers[0]
    for n in numbers[1:]:
        if n < min_val:
            min_val = n
    return min_val


def math_operations():
    """
    Interactive program to perform mathematical operations on user-provided numbers.
    Handles invalid input and empty lists using try-except.

    Returns:
        float or None: The result of the chosen operation, or None on error.
    """
    print("=== Math Operations ===")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Choose operation (1/2/3/4): ").strip()
        raw = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw.split()]

        operations = {
            '1': (compute_sum,     'Sum'),
            '2': (compute_average, 'Average'),
            '3': (compute_maximum, 'Maximum'),
            '4': (compute_minimum, 'Minimum'),
        }

        if choice not in operations:
            print("Invalid choice. Please select 1–4.")
            return None

        func, label = operations[choice]
        result = func(numbers)
        print(f"\n{label}: {result}")
        return result

    except ValueError as e:
        print(f"Error: {e}")
        return None

math_operations()
```

## 4.2 Exercise on List Manipulation

```python
# 1. Extract Every Other Element
def extract_every_other(lst):
    """
    Returns a new list containing every other element starting from index 0.

    Parameters:
        lst (list): The input list.

    Returns:
        list: Elements at indices 0, 2, 4, ...
    """
    return lst[::2]


# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """
    Returns a sublist from start index to end index (inclusive).

    Parameters:
        lst (list): The input list.
        start (int): Starting index (0-based).
        end (int): Ending index (inclusive).

    Returns:
        list: The sublist from start to end inclusive.
    """
    return lst[start:end + 1]


# 3. Reverse a List Using Slicing
def reverse_list(lst):
    """
    Returns the input list in reverse order using slicing.

    Parameters:
        lst (list): The input list.

    Returns:
        list: Reversed version of lst.
    """
    return lst[::-1]


# 4. Remove First and Last Elements
def remove_first_last(lst):
    """
    Returns the list without its first and last elements.

    Parameters:
        lst (list): The input list (must have at least 2 elements).

    Returns:
        list: Sublist without the first and last elements.
    """
    return lst[1:-1]


# 5. Get First n Elements
def get_first_n(lst, n):
    """
    Returns the first n elements of a list.

    Parameters:
        lst (list): The input list.
        n (int): Number of elements to extract from the start.

    Returns:
        list: First n elements of lst.
    """
    return lst[:n]


# 6. Extract Last n Elements
def get_last_n(lst, n):
    """
    Returns the last n elements of a list.

    Parameters:
        lst (list): The input list.
        n (int): Number of elements to extract from the end.

    Returns:
        list: Last n elements of lst.
    """
    return lst[-n:]


# 7. Extract Elements in Reverse Order (skip one, start from second-to-last)
def reverse_skip(lst):
    """
    Returns every second element starting from the second-to-last, moving backward.

    Parameters:
        lst (list): The input list.

    Returns:
        list: Every other element from the second-to-last to the start.
    """
    return lst[-2::-2]


lst1 = [1, 2, 3, 4, 5, 6]

print("extract_every_other: ", extract_every_other(lst1))
print("get_sublist: ", get_sublist(lst1, 2, 4))
print("reverse_list: ", reverse_list(lst1))
print("remove_first_last: ", remove_first_last(lst1))
print("get_first_n: ", get_first_n(lst1, 3))
print("get_last_n: ", get_last_n(lst1, 2))
print("reverse_skip: ", reverse_skip(lst1))
```

## 4.3 Exercise on Nested Lists

```python
# 1. Flatten a Nested List (one level)
def flatten(lst):
    """
    Flattens a singly-nested list into a single flat list.

    Parameters:
        lst (list of lists): A nested list (one level deep).

    Returns:
        list: A flat list with all elements from inner lists.
    """
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result


# 2. Access Nested Element by Indices
def access_nested_element(lst, indices):
    """
    Retrieves an element from a nested list given a sequence of indices.

    Parameters:
        lst (list): A nested list.
        indices (list of int): A list of indices specifying the path to the element.

    Returns:
        The element found at the specified nested position.
    """
    element = lst
    for idx in indices:
        element = element[idx]
    return element


# 3. Sum of All Elements in a Nested List (any depth)
def sum_nested(lst):
    """
    Recursively computes the sum of all numeric elements in a nested list.

    Parameters:
        lst (list): A nested list of numbers at any depth.

    Returns:
        float: The total sum of all elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# 4. Remove Specific Element from Nested List
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from each inner list.

    Parameters:
        lst (list of lists): A nested list.
        elem: The element to remove from every inner list.

    Returns:
        list: The nested list with all occurrences of elem removed.
    """
    return [[x for x in sublist if x != elem] for sublist in lst]


# 5. Find Maximum Element in a Nested List (any depth)
def find_max(lst):
    """
    Recursively finds the maximum element in a nested list.

    Parameters:
        lst (list): A nested list of numbers at any depth.

    Returns:
        float: The maximum numeric value found.
    """
    max_val = None
    for item in lst:
        if isinstance(item, list):
            candidate = find_max(item)
        else:
            candidate = item
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val


# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
        lst (list): A nested list.
        elem: The element to count.

    Returns:
        int: The total number of occurrences of elem.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# 7. Deep Flatten (any depth)
def deep_flatten(lst):
    """
    Recursively flattens a deeply nested list into a single flat list.

    Parameters:
        lst (list): A nested list of arbitrary depth.

    Returns:
        list: A completely flat list containing all elements.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# 8. Average of All Elements in a Nested List
def average_nested(lst):
    """
    Computes the average of all elements in a nested list.

    Parameters:
        lst (list): A nested list of numbers.

    Returns:
        float: The arithmetic mean of all elements.
    """
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)



print("flatten ([[1,2],[3,4],[5]]) :", flatten([[1, 2], [3, 4], [5]]))

nested_3x3 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("access_nested_element ([[1,2,3],[4,5,6],[7,8,9]], [1,2]) : ", access_nested_element(nested_3x3, [1, 2]))
print("")
print("sum_nested ([[1,2],[3,[4,5]],6]) : ", sum_nested([[1, 2], [3, [4, 5]], 6]))
print("")
print("remove_element ([[1,2],[3,2],[4,5]], 2) : ", remove_element([[1, 2], [3, 2], [4, 5]], 2))
print("")
print("find_max ([[1,2],[3,[4,5]],6]) : ", find_max([[1, 2], [3, [4, 5]], 6]))
print("")
print("count_occurrences([[1,2],[2,3],[2,4]], 2): ", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
print("")
print("deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]): ", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
print("")
print("average_nested([[1,2],[3,4],[5,6]]): ", average_nested([[1, 2], [3, 4], [5, 6]]))
```

## 10.2 Experiment — How Fast is NumPy?

```python
import time
import numpy as np

SIZE = 1_000_000

# Prepare data
list_a = list(range(SIZE))
list_b = list(range(SIZE))
np_a   = np.arange(SIZE, dtype=float)
np_b   = np.arange(SIZE, dtype=float)


# 1. Element-wise Addition
t0 = time.time()
result_list = [list_a[i] + list_b[i] for i in range(SIZE)]
t_list_add = time.time() - t0

t0 = time.time()
result_np = np_a + np_b
t_np_add = time.time() - t0

print("=" * 55)
print(f"1. Element-wise Addition")
print(f"   Python list : {t_list_add:.5f} s")
print(f"   NumPy array : {t_np_add:.5f} s")
print(f"   Speedup     : {t_list_add / t_np_add:.1f}x")


# 2. Element-wise Multiplication
t0 = time.time()
result_list = [list_a[i] * list_b[i] for i in range(SIZE)]
t_list_mul = time.time() - t0

t0 = time.time()
result_np = np_a * np_b
t_np_mul = time.time() - t0

print(f"\n2. Element-wise Multiplication")
print(f"   Python list : {t_list_mul:.5f} s")
print(f"   NumPy array : {t_np_mul:.5f} s")
print(f"   Speedup     : {t_list_mul / t_np_mul:.1f}x")


#  3. Dot Product
t0 = time.time()
dot_list = sum(list_a[i] * list_b[i] for i in range(SIZE))
t_list_dot = time.time() - t0

t0 = time.time()
dot_np = np.dot(np_a, np_b)
t_np_dot = time.time() - t0

print(f"\n3. Dot Product")
print(f"   Python list : {t_list_dot:.5f} s")
print(f"   NumPy array : {t_np_dot:.5f} s")
print(f"   Speedup     : {t_list_dot / t_np_dot:.1f}x")


# 4. Matrix Multiplication (1000x1000)
MAT_SIZE = 1000

# Python list matrix multiplication (naive O(n³))
mat_a = [[float(i * MAT_SIZE + j) for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]
mat_b = [[float(i * MAT_SIZE + j) for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]

print(f"\n4. Matrix Multiplication ({MAT_SIZE}x{MAT_SIZE})")
print("   Running Python list matrix multiplication (this may take a moment)...")

t0 = time.time()
# Transpose mat_b once to speed up column access
mat_b_T = [[mat_b[j][i] for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]
result_mat = [
    [sum(mat_a[r][k] * mat_b_T[c][k] for k in range(MAT_SIZE)) for c in range(MAT_SIZE)]
    for r in range(MAT_SIZE)
]
t_list_mat = time.time() - t0

np_mat_a = np.array(mat_a)
np_mat_b = np.array(mat_b)
t0 = time.time()
result_np_mat = np_mat_a @ np_mat_b
t_np_mat = time.time() - t0

print(f"   Python list : {t_list_mat:.5f} s")
print(f"   NumPy array : {t_np_mat:.5f} s")
print(f"   Speedup     : {t_list_mat / t_np_mat:.1f}x")
print("=" * 55)
print("Conclusion: NumPy is dramatically faster for numerical")
print("operations, especially as array size increases, because")
print("it uses optimised C/Fortran code under the hood.")
```
